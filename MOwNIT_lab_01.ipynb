```python
import numpy as np
from tabulate import tabulate

f16 = np.finfo(np.float16)
f32 = np.finfo(np.float32)
f64 = np.finfo(np.float64)
f128 = np.finfo(np.float128)

table = tabulate([['float 16', f16.precision,
                  f16.min, f16.max ],
                  ['float 32', f32.precision,
                  f32.min, f32.max],
                  ['float 64', f64.precision,
                  f64.min, f64.max],
                  ['float 128', f128.precision,
                  f128.min, f128.max]],
                 headers=['Type', 'Precision', 'Min value', 'Max value'])
print(table)

```

### Zadanie 2 
Napisz funkcje, która dla dowolnej liczby zmiennoprzecinkowej w precyzji 64-bitowej (w standardzie IEEE 754) przedstawi w wersji dziesiętnej i binarnej składniki liczby: znak, wykładnik i mantysa. Można wykorzystać między innymi bibliotekę bitstring. Zbadaj jak reprezentowaną są poszczególne składniki tego formatu: zakres, sposób kodowania liczby etc. 

```python
import bitstring as bs
def number_parts(number):
    float64 = bs.BitArray(float=number, length=64)
    exponent = float64.bin[1:12]
    mantissa = float64.bin[12:64]

    if float64[0] == 0:
        dec_sign = '+'
    else:
        dec_sign = '-'

    dec_exponent = 0
    pow = 10
    for i in exponent[0:]:
        if i == '1':
            dec_exponent += 2 ** pow
        pow -= 1

    dec_mantissa = 0
    pow = -1
    for i in mantissa[0:]:
        if i == '1':
            dec_mantissa += 2 ** pow
        pow -= 1

    print('Sign/Exponent/Mantissa',float64.bin[0], exponent, mantissa)
    print('Sign/Exponent/Mantissa',dec_sign, dec_exponent, dec_mantissa)

number_parts(7.25)
```
