```python
import numpy as np

print("Check if vector contains infinity value/values:")
X_vector = np.array([-np.inf, 5,3,4,np.inf, 21])
Y_vector = np.array([0,0,0,2,3,4,1,6])

#Identifying infinity
#Variant I
def does_contain_inf1(array):
    for i in np.isfinite(array):
        if i == False:
            print("There is infinity value in given vector")
            return True
    print("Ther is no infinity values in given vector")
    return False

print("\nVariant I:")
does_contain_inf1(X_vector)
does_contain_inf1(Y_vector)

#Variant II
def does_contain_inf2(array):
    for i in np.isinf(array):
        if i == True:
            print("There is infinity value in given vector")
            return True
    print("Ther is no infinity values in given vector")
    return False

print("\nVariant II:")
does_contain_inf2(X_vector)
does_contain_inf2(Y_vector)

#Variant III
def does_contain_inf3(array):
    for i, j in zip(np.isposinf(array), np.isposinf(array)):
        if i == True or j == True:
            print("There is infinity value in given vector")
            return True
    print("Ther is no infinity values in given vector")
    return False

print("\nVariant III:")
does_contain_inf3(X_vector)
does_contain_inf3(Y_vector)


#Identifying zeros

print("\nCheck if vector contains zero value/values: \n ")
print("Varraint I:")
def does_contain_zero1(array):
    if np.count_nonzero(array)==len(array):
        print("There is no zero value in given vector")
        return False
    print("There is zero value in given vector")
    return True

does_contain_zero1(X_vector)
does_contain_zero1(Y_vector)


print("\nVarraint II:")
def does_contain_zero2(array):
    if len(np.flatnonzero(array))==len(array):
        print("There is no zero value in given vector")
        return False
    print("There is zero value in given vector")
    return True

does_contain_zero2(X_vector)
does_contain_zero2(Y_vector)

print("\n Variant III:")
def does_contain_zero3(array):
    if np.nonzero(array)[0].size==len(array):
        print("There is no zero value in given vector")
        return False
    print("There is zero value in given vector")
    return True

does_contain_zero3(X_vector)
does_contain_zero3(Y_vector)

#Comparing matrix I

def compare_matrix_or_vectors1(m1,m2):
    if np.equal(m1,m2).all():
        print("Given matrix/vectors are equal")
    elif np.greater(m1,m2).all():
        print("Values of matrix/vector given as parameter one are greater than values of matrix/vector given as parameter 2")
    elif np.greater_equal(m1,m2).all():
        print("Values of matrix/vector given as parameter one are greater or equal than values of matrix/vector given as parameter 2")
    elif np.less(m1,m2).all():
        print("Values of matrix/vector given as parameter 1 are smaller than values of matrix/vector given as parameter 2")
    elif np.less_equal(m1,m2).all():
        print("Values of matrix/vector given as parameter 1 are smaller or equal than values of matrix/vector given as parameter 2")
    else :
        print("There is no relation between values of given matrix/vector")

#Comparing matrix II


def compare_matrix_or_vectors2(m1, m2):
    if m1.ndim == m2.ndim and m1.size == m2.size:
        return m1 == m2
    elif m1.size > m2.size:
        m2 = np.resize(m2, (1, m1.size))
        m1 = np.resize(m1, (1, m1.size))
        return m1 == m2
    else:
        m1 = np.resize(m2, (1, m2.size))
        m2 = np.resize(m2, (1, m2.size))
        return m1 == m2


m1 = np.array([1, 2, 3, 4, 5])
m2 = np.array([1, 2, 5])
m3 = np.array([[1, 2], [3, 4]])
vec1 = np.array([1,2,3,4])
vec2 = np.array([2,3,4,5])
matrix1 = np.array([[1,1,1], [2,2,2], [3,3,3]])
matrix2 = np.array([[1,1,1], [2,2,2], [3,3,3]])

print("\nVariant I:")
print("\nComparing matrix, method appropriate only for matrix of the same size:")
compare_matrix_or_vectors1(vec2,vec1)
compare_matrix_or_vectors1(vec1,vec2)
compare_matrix_or_vectors1(matrix1,matrix2)

print("\nVariant II: \n")
print("Comparing matrix.:\n")
print(compare_matrix_or_vectors2(m1, m2), '\n')
print(compare_matrix_or_vectors2(m3,m1), '\n')

def are_equal(a,b):
  print("Are matrix equal?", np.array_equal(a,b))
print("\nVariant III: \n")
are_equal(matrix1, matrix2);
are_equal(m1, m2);
```
